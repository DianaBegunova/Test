## Тестовое задание 
На основе уже знакомого тестового набора данных из первого задания посчитать следующие метрики в тетрадке Jupyter Notebook:
- Delivery rate
- Open rate
- Click to Open rate
- Unsubscribe rate
- Выяви лучшую тему
- И лучший день недели (для планирования рассылок)

## Посмотрим данные

```python
# подгрузим библиотеки
import pandas as pd
```

```python
# прочитаем данные
data = pd.read_excel('test\Data для тестового.xlsx')
pd.set_option('display.max_columns', None)

# ознакомимся с данными
data.info()
data.head()
```

### Метрики

```python
# рассчитаем метрики
data['delivery_rate'] = (data['Доставлено'] / data['Отправлено']) * 100
data['open_rate'] = (data['Открытия'] / data['Доставлено']) * 100
data['click_to_open_rate'] = (data['Клики'] / data['Открытия']) * 100
data['unsubscribe_rate'] = (data['Отписки'] / data['Доставлено']) * 100

# выведем первые строки данных с добавленными метриками
data.head()
```

## Лучшая тема писем

```python
# выведем темы писем и соответствующие метрики
theme_metrics = data[['Тема письма ', 'open_rate']]
theme_metrics = theme_metrics.sort_values(by='open_rate', ascending=False)
theme_metrics.head()
```

В лидерах оказались 2 темы: 1 и 101. Посмотрим другие метрики, чтобы определиться с единственной и неповторимой темой.

```python
# создаем срез для темы письма "Тема письма 1"
theme_slice = data[(data['Тема письма '] == "Тема письма 1") | (data['Тема письма '] == "Тема письма 101")]

# выводим только необходимые столбцы
selected_columns = theme_slice[['Тема письма ', 'delivery_rate', 'open_rate', 'click_to_open_rate', 'unsubscribe_rate']]
selected_columns
```

Метрики у этих двух тем схожи, однако click_to_open_rate различается в сторону 1 темы, поэтому:
**Лучшая тема письма = тема 1**

## Лучший день недели (для планирования рассылок)

```python
# группируем данные по дню недели и смотрим средние значения метрик
day_of_week_metrics = data.groupby('День недели')[['delivery_rate', 'open_rate', 'click_to_open_rate', 'unsubscribe_rate']].mean().round(2)

# выведем результаты
day_of_week_metrics
```

Нам по большей части важны высокие показатели у Open Rate и Click to Open Rate, и низкие у Unsubscribe Rate. В таком случае сделаем основание на 1 критерии с учетом не проседа во 2 и 3 критериях.

```python
def find_best_day(data):
    
    # сортируем дни недели по Open Rate (по убыванию)
    sorted_days = day_of_week_metrics.sort_values(by='open_rate', ascending=False)
    
    # выбирем первый день с учетом критериев Click to Open Rate и Unsubscribe Rate
    best_day = sorted_days[(sorted_days['click_to_open_rate'] != sorted_days['click_to_open_rate'].min()) &
                           (sorted_days['unsubscribe_rate'] != sorted_days['unsubscribe_rate'].max())].index[0]

    return best_day

# применим функцию к данным
best_day_of_week = find_best_day(data)

# выведем результат
print("Лучший день недели:", best_day_of_week)
```

## Вывод

- **Лучшая тема письма - 1.**
- **Лучший день недели - 3 (= среда).**
